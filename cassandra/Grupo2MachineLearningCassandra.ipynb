```python
import pandas as pd
from sklearn import tree
from sklearn import model_selection
```

```python
dfDatosBasicos = pd.read_csv('CSV_Modificado.csv')
```

```python
dfDatosBasicos = dfDatosBasicos.drop(['IncidntNum', 'Descript', 'Resolution', 'Address', 'Location', 
              'PdId'], 1)
```

```python
dfDatosBasicos['Datetime'] = pd.to_datetime(dfDatosBasicos['Date'] + ' ' + dfDatosBasicos['Time'])
dfDatosBasicos = dfDatosBasicos.drop(['Date', 'Time'], 1)

dfDatosBasicos['Hour'] = dfDatosBasicos['Datetime'].apply(lambda d: d.hour + d.minute / 60)
dfDatosBasicos['DayOfWeek'] = dfDatosBasicos['Datetime'].apply(lambda d: d.weekday())
dfDatosBasicos['DayOfMonth'] = dfDatosBasicos['Datetime'].apply(lambda d: d.day)
dfDatosBasicos['Month'] = dfDatosBasicos['Datetime'].apply(lambda d: d.month)
dfDatosBasicos['Year'] = dfDatosBasicos['Datetime'].apply(lambda d: d.year)
dfDatosBasicos = dfDatosBasicos.drop('Datetime', 1)
```

```python
dfDatosBasicos = pd.concat([dfDatosBasicos, pd.get_dummies(dfDatosBasicos['PdDistrict'])],1)
dfDatosBasicos = dfDatosBasicos.drop('PdDistrict',1)
dfDatosBasicos.head(10)
```

```python
catValues = dfDatosBasicos.drop('Category',1).values
labels= dfDatosBasicos['Category'].values
```

```python
clasificador = tree.DecisionTreeClassifier()
kfold = model_selection.KFold(n_splits = 10)
resultados = model_selection.cross_val_score(clasificador, catValues, labels, cv=kfold,scoring='accuracy')
print('Precision: %f (%f)' % (resultados.mean(), resultados.std()))
```
